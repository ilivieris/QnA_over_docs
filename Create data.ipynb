```python
from goose3.goose3 import Goose
```

```python
g = Goose()

visited_sites = []
L = ['Napoleon']

while (L != []):
    site = L.pop(-1)
    if (site in visited_sites): continue
    else: visited_sites.append(site)

    if ('.jpg' in site): continue
    if (':' in site): continue

    url = f'https://en.wikipedia.org/wiki/{site}'
    article = g.extract(url)
    
    
    if (article.infos['opengraph']['type'] != 'website'): continue
    if ('Look foron one of Wikipedia' in article.cleaned_text): continue
    print(f'Site: {site}\t\t List length: {len(L)}')

    # TODO
    with open(f"Data/{site}.txt","w") as f:
        f.write(article.cleaned_text)

    for link in article.links:
        if '/wiki/' in link: 
            if (link.split('/')[-1] not in visited_sites): L.append(link.split('/')[-1])
```
